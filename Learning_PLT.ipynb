```python
import matplotlib.pyplot as plt
names = ['group_a', 'group_b', 'group_c']
values = [1,10,100]
plt.figure(figsize=(45,5))
subtitle_string = "Bar Chart"
plt.subplot(151)
plt.bar(names,values)
plt.title(subtitle_string,fontsize=20)
subtitle_string = "Pie Chart"
plt.subplot(152)
plt.pie(values)
plt.title(subtitle_string,fontsize = 20)
plt.subplot(153)
plt.plot(names, values)
plt.xlabel("Yeah bitch", color = "red", fontsize = 50)
plt.suptitle("I Fucking Did It?", fontsize = 50, color = "red")
plt.show()
```
